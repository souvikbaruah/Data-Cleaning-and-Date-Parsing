From the above output we can clearly see that the columns 'DOB', 'Arrival' and 'Departure' have dates stored in them. In order to verify that the values stored by these columns are dates we have to check whether the datatype of these three columns is 'object'. The reason for this is explained below after checking the datatypes of the three columns.

```
# printing the first few rows of the columns 'DOB', 'Arrival' and 'Departure'
print(data['DOB'].head())
print(data['Arrival'].head())
print(data['Departure'].head())
```

We can see that the datatypes of the columns 'DOB', 'Arrival' and 'Departure' have been shown as 'object'. Pandas uses this datatype to refer mainly to strings. As dates are nothing but strings, the output 'dtype: object' verifies that these columns have dates stored in them.

```
# checking the data type of 'DOB' column without printing the first few rows
data['DOB'].dtype
```

```
# checking the data type of 'Arrival' column without printing the first few rows
data['Arrival'].dtype
```

```
# checking the data type of 'Departure' column without printing the first few rows
data['Departure'].dtype
```

We can see that the datatypes of the columns 'DOB', 'Arrival' and 'Departure' are shown as 'O' where 'O' is nothing but the code for 'object'.

```
#Since, Pandas doesn't automatically recognise the date values as 'date', therefore, we have to convert them so that they are recognized as dates.
# Creating a new column, DOB_1, with the parsed dates 
# Handling different date formats present in the dataset and converting them into the same format 'year-month-day'
data['DOB_1'] = pd.to_datetime(data['DOB'], infer_datetime_format=True)
data['DOB_1'].head()
```

```
# creating a new column, Arrival_1, with the parsed dates
# Handling different date formats present in the dataset and converting them into the same format 'year-month-day'
data['Arrival_1'] = pd.to_datetime(data['Arrival'], infer_datetime_format=True)
data['Arrival_1'].head()
```

```
# creating a new column, Departure_1, with the parsed dates
# Handling different date formats present in the dataset and converting them into the same format 'year-month-day'
data['Departure_1'] = pd.to_datetime(data['Departure'], infer_datetime_format=True)
data['Departure_1'].head()
```

We can clearly see that now when we check the datatypes of our new columns 'DOB_1', 'Arrival_1' and 'Departure_1' we get the result as 'datetime64'. They are now specifically recognized as dates.

```
# printing first few rows of our dataset after parsing the dates
data.head()
```

After this, we print the difference of our newly parsed date columns taking two at a time.

```
# storing the difference between 'DOB_1' and 'Arrival_1' in a new column name 'DOB_1-Arrival_1'
data['DOB_1-Arrival_1'] = (data['DOB_1'].sub(data['Arrival_1'], axis=0)).abs()
data.head()
```

```
# storing the difference between 'Arrival_1' and 'Departure_1' in a new column name 'Arrival_1-Departure_1'
data['Arrival_1-Departure_1'] = (data['Arrival_1'].sub(data['Departure_1'], axis=0)).abs()
data.head()
```

```
# storing the difference between 'Departure_1' and 'DOB_1' in a new column name 'Departure_1-DOB_1'
data['Departure_1-DOB_1'] = (data['Departure_1'].sub(data['DOB_1'], axis=0)).abs()
data.head()
```

```
# loading the dataset 'pokemon.csv' (publicly available dataset collected from Kaggle) stored in the 'task' folder of Google Drive into 'data'
data=pd.read_csv("/content/drive/My Drive/task/pokemon.csv")
#displaying first few rows of our dataset 
data.head()
```

```
#checking the number of rows and columns present in our dataset
data.shape
```

We see that our dataset contains a total of 721 rows and 23 columns. Now, the task is to drop all the columns having Pearson correlation more than 0.85.

```
# Creating the correlation matrix
corr_matrix = data.corr().abs()

# Selecting the upper triangle of the correlation matrix
upper = corr_matrix.where(np.triu(np.ones(corr_matrix.shape), k=1).astype(np.bool))

# Finding the index of the feature columns with Pearson correlation greater than 0.85
to_drop = [column for column in upper.columns if any(upper[column] > 0.85)]
```

```
# Droping the columns having Pearson correlation more than 0.85
# after dropping, we print the dataset again to verify that some of the columns are dropped
data.drop(data[to_drop], axis=1)
```

So, we can see that the 'Generation' column has been dropped and the total number of columns got reduced to 22 !
